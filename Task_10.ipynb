```python
import random
import pandas as pd

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df.head(10)
```

```python
pd.get_dummies(df, columns=['whoAmI'])
```

```python
one_hot_data = pd.DataFrame({'human': [1 if i == 'human' else 0 for i in lst],
                             'robot': [1 if j == 'robot' else 0 for j in lst]})
```
